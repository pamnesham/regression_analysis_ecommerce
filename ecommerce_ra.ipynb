# App vs Website Performance: E-Commerce Data Analysis



The following analysis uses the "Ecommerce Customers" dataset from Pierian Data. The goal of this analysis is to determine if this ecommerce company should focus efforts on its app or its website in order to increase the Yearly Amount Spent per customer.


```python
# import basic packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read the data into python
customers = pd.read_csv("Ecommerce Customers")
```

## Training the Data and Assesing the Model
This part of the analysis splits the data into testing and training data and then uses a linear regression model to fit the data for prediction.

First let's grab the independent variable from the column names and assign it to y, then we'll grab the dependendent variables and assign them to X.

```python
customers.columns
```

```python
y = customers['Yearly Amount Spent']
X = customers[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

Now we split the data into training and testing groups with .3 of the data being set aside for testing. We'll do this using train_test_split from sklearn. 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Now we'll use sklearn's LinearRegression module and .fit method to fit our  training data on a linear regression model. The .fit method determines our data's coefficients and intercept. It basically creates the functon or equation that shows the best line of fit in the data. 

```python
from sklearn.linear_model import LinearRegression
```

```python
# create the LinearRegression object
lm = LinearRegression()
```

```python
# fit the training data to the linear regression model
lm.fit(X_train,y_train)
```

Once the data has been trained on the model, we'll see if we can predict the test y using our test X. Since a variable will perfectly correlate with itself, we would expect our training y to be very close to perfectly closely correlated with our test y if our model is a good fit for the data.

```python
predicted_y = lm.predict(X_test) # use the untrained X data to see if it can predict the untrained y
```

```python
plt.scatter(y_test, predicted_y)
plt.xlabel('y_test')
plt.ylabel('predicted_y')
```

That looks like a pretty strong correlation. 

To be sure that this is a good model it's important to calculate error to see how accurate our predictions were.

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predicted_y))
print('MSE:', metrics.mean_squared_error(y_test, predicted_y))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predicted_y)))
```

This appears to be an acceptably small amount of error for our data. 

Another way to ensure we have chosen the best predictive model for our data is to graph the distribution of the error. We would expect our error to be normally distributed due to the central limit theorem that states that given a set of random independent variables we would expect those variables to be normally distributed. If the error is not normally distributed, this is a good sign that the error is caused by not having the best fitting model.

```python
# take the actual y values and subtract the predictions to find the error
# a normal distribution indicates a correct model choice
sns.distplot(y_test-predicted_y)
```

Since our error is normally distributed we can conclude that our model is  a good fit. We can also calculate the explained variance score to show that our model is explaining 98.9%  of the variance in the data:

```python
metrics.explained_variance_score(y_test, predicted_y)
```
